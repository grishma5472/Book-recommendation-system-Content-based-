```python
#since all these books are 'Goodreads Best' and since we are doing CONTENT-BASED classification, we do not need ratings and number of ratings.
books=books[['Book','Author','Description','Genres','URL']]
books.head()
```

```python
#for tags, let us merge author, description, genres
books.isnull().sum()
```

```python
books.dropna(inplace=True)
```

```python
books.duplicated().sum()
```

```python
books['Genres'][0]
```

```python
books['Description']=books['Description'].apply(lambda x:x.split())
```

```python
import ast
def convert_to_list(text):
  return ast.literal_eval(text)
```

```python
books['Genres']=books['Genres'].apply(convert_to_list)
```

```python
def make_list(text):
  return [text]
```

```python
books['Author']=books['Author'].apply(make_list)
books.head()
```

```python
books['Genres']=books['Genres'].apply(lambda x:[i.replace(" ","") for i in x])
books['Author']=books['Author'].apply(lambda x:[i.replace(" ","") for i in x])
books.head()
```

```python
books['tags']=books['Description']+books['Genres']+books['Author']
books.head() #merging the lists
```

```python
final=books[['Book','tags','URL']]
final.head()
```

```python
final['tags']=final['tags'].apply(lambda x:" ".join(x)) #converting list to string
```

```python
final['tags'][100] #based on the tag, this is definitely the Da Vinci Code, one of my favs :)))
```

```python
#Stemming the words to make sure words like, liked, liking represent the same thing
import nltk
from nltk.stem.porter import PorterStemmer
port=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(port.stem(i))

  return " ".join(y)
```

```python
final['tags']=final['tags'].apply(stem)
```

```python
#let us convert the string to lowercase
final['tags']=final['tags'].apply(lambda x:x.lower())
```

```python
final.head()
```

```python
#We gotta use these tags to find similarity among books. Convert these strings to vectors
#BAG OF WORDS
#we do not consider stop words
#the distances among the vectors are calculated which becomes an indicator of similarity among books

from sklearn.feature_extraction.text import CountVectorizer
model=CountVectorizer(max_features=5000,stop_words='english')
vectors=model.fit_transform(final['tags']).toarray()
```

```python
#We have to calculate the distance of every book with the other books
#We will calculate the cosine distance (not the euclidean distance)
#In higher dimensions, we use cosine similarity(inversely proportional to cosine distance) not euclidean
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
similarity.shape
```

```python
similarity
```

```python
# on this similarity array of lists, we need to enumerate to make pairs
#so that index is also stored along with the value
```

```python
#All the diagonal elements will be 1 because that is the position of that book itself
def fetch_recs(movie):
  index=final[final['Book']==movie].index[0]
  distances=similarity[index]
  book_recs=sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[0:16]
  for i in book_recs:
    print(final.iloc[i[0]].Book)
```
